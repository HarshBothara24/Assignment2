```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Housing.csv')
df.head()
```

```python
sns.scatterplot(x='area', y='price', data=df)
plt.title('Price vs Area')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.show()
```

```python
plt.hist(df['price'])
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='bedrooms', y='price', data=df, palette='Set2', hue='bedrooms')
plt.title('Price Distribution by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')

plt.show()
```

```python
sns.barplot(x='stories', y='price', data=df, estimator=np.mean)
plt.title('Average Price by Number of Stories')
plt.xlabel('Number of Stories')
plt.ylabel('Average Price')
plt.show()
```

```python
sns.histplot(df['price'], bins=20, kde=True)
plt.title('Distribution of Property Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
sns.pairplot(df)
plt.title('Pairplot of Real Estate Data')
plt.show()
```

```python
sns.countplot(x='bedrooms', data=df)
plt.title('Count of Properties by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.show()
```

```python
sns.lineplot(x='area', y='price', data=df)
plt.title('Price Trend by Area')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.show()
```

```python
g = sns.FacetGrid(df, col="stories")
g.map(sns.histplot, "price")
plt.show()
```
